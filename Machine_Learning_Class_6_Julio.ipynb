# Metricas

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons #funcion cuadratica (curva en el eje de las x y y)
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import (accuracy_score, confusion_matrix, precision_score,
                             recall_score, f1_score, classification_report)
```

```python
x, y = make_moons(n_samples=(150,50))
```

# Morado = x
# Amarillo = Y

```python
# si queremos ver los datos
plt.scatter(x[:,0],x[:,1],c=y) #seleccionamos todos los valores que hay en los raws, en los datos
plt.grid()
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
clf=MLPClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
#Precision
clf.score(x_train,y_train)
```

# hasta cuando lo vamos a bajar, comprobar que los datos son reales
## sacamos las predicciones
## variables que vayamos a predecir

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
accuracy= accuracy_score(y_test,y_pred)
accuracy
```

```python
matriz=confusion_matrix(y_test, y_pred)
matriz
```

```python
precision = precision_score(y_test, y_pred)
precision
```

```python
recall = recall_score (y_test,y_pred)
recall
```

## Esta Red Neuronal al parecer si nos puede dar un resultado

```python
f1=f1_score(y_test,y_pred)
f1
```

# Se puede aplicar otro método conocido como Neighbors

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons #funcion cuadratica (curva en el eje de las x y y)
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (accuracy_score, confusion_matrix, precision_score,
                             recall_score, f1_score, classification_report)
```

```python
x, y = make_moons(n_samples=(150,50))
```

```python

plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
clf=KNeighborsClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
#Precision
clf.score(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
accuracy= accuracy_score(y_test,y_pred)
accuracy
```

```python
matriz=confusion_matrix(y_test, y_pred)
matriz
```

```python
precision = precision_score(y_test, y_pred)
precision
```

```python
recall = recall_score (y_test,y_pred)
recall
```

```python
f1=f1_score(y_test,y_pred)
f1
```

# Del mismo modo con Gaussian Process Classifier

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons #funcion cuadratica (curva en el eje de las x y y)
from sklearn.model_selection import train_test_split
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.metrics import (accuracy_score, confusion_matrix, precision_score,
                             recall_score, f1_score, classification_report)
```

```python
x, y = make_moons(n_samples=(150,50))
```

```python

plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
clf=GaussianProcessClassifier()
```

```python
clf.fit(x_train,y_train)
```

```python
#Precision
clf.score(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
accuracy= accuracy_score(y_test,y_pred)
accuracy
```

```python
matriz=confusion_matrix(y_test, y_pred)
matriz
```

```python
precision = precision_score(y_test, y_pred)
precision
```

```python
recall = recall_score (y_test,y_pred)
recall
```

```python
f1=f1_score(y_test,y_pred)
f1
```
